```python
import matplotlib.pyplot as plt
import numpy as np

# 
np.random.seed(0)
# data = np.random.rand(10, 12)
data_ori = np.array([
    [77.80, 74.79, 68.98, 54.37, 49.94, 36.32], 
    [17.24, 24.58, 20.38, 16.58, 15.77, 11.85], 
    [5.63, 8.09, 8.51, 8.23, 8.39, 7.34], 
    [33.55, 35.65, 37.64, 35.06, 34.35, 32.56]
])
data = np.array([
    [33.55, 35.65, 37.64, 35.06, 34.35, 32.56], 
    [5.63, 8.09, 8.51, 8.23, 8.39, 7.34],
    [17.24, 24.58, 20.38, 16.58, 15.77, 11.85],
    [77.80, 74.79, 68.98, 54.37, 49.94, 36.32]
])
print("data", data)

# 
fig, ax = plt.subplots(figsize=(10, 8))

# 
cax = ax.matshow(data, cmap='Blues') # cmap='coolwarm'

# 
fig.colorbar(cax)

# 
ax.set_xticks(np.arange(data.shape[1]))
ax.set_yticks(np.arange(data.shape[0]))
ax.set_xticklabels([f'Col {i}' for i in range(6)])
ax.set_yticklabels([f'Row {i}' for i in range(4)])

# 
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        text = ax.text(j, i, f'{data[i, j]:.2f}', ha='center', va='center', color='black')

# 
plt.title('Heatmap Example')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.savefig("heatmap.svg", format="svg")

# 
plt.show()
```
